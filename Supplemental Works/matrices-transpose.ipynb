```
import numpy as np
```

Matrices as numpy arrays

```
A = np.array([[-1,0,2],[5,-3,0]])
print(A)
```

$A[i]$ gives the $(i+1)^{th}$ row

```
A[0] #this is the first row
```

Two ways to the components of matrices

```
print(A[0][2])
print(A[0,2])
```

```
print(A)
x = 3
print(x*A)
```

```
A = np.array([[-1,0,2],[5,-3,0]])
B = np.array([[1,10,12],[7,-1,2]])
print(A)
print(B)
print(A+B)
```

Matrix-matrix multiplication

```
A = np.array([[2,-1],[0,1]])
B = np.array([[-1,1,2],[1,0,3]])
np.matmul(A,B)
```
