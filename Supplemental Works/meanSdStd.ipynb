Compute the standard deviation

```
mu = np.mean(x)
s = 0
for a in x:
  s = s + (a-mu)**2
s = s/len(x)
s = np.sqrt(s)
print(s)  
```

```
st = np.std(x)
print(st)
```

```
plt.plot(x,np.zeros(len(x)),'ko',markersize=6)
plt.plot([mu-st,mu+st],[0,0],'go',markersize=6)
plt.plot(mu,0,'ro',markersize=6)
```

```
x = np.array([3.001,2.998,2.997,3.002])
mu = np.mean(x)
st = np.std(x)
plt.plot(x,np.zeros(len(x)),'ko',markersize=6)
plt.plot([mu-st,mu+st],[0,0],'go',markersize=6)
plt.plot(mu,0,'ro',markersize=6)
print(mu,st)

```

```
x_scaled = (x-mu)/st
print(np.mean(x_scaled))
print(np.std(x_scaled))
plt.plot(x_scaled,np.zeros(len(x_scaled)),'ko',markersize=6)
plt.plot([-1,1],[0,0],'go',markersize=6)
plt.plot(0,0,'ro',markersize=6)

```

```
# input: array x (1 or 2 dimensional)
# output: scaled x, mean of x, standard deviation of x
# note: scaled x = (x - mean of x)/standard deviation of x   
def scaled_array(x):
  mu = np.mean(x,axis=0)
  st = np.std(x,axis=0)
  return (x-mu)/st,mu,st
```

```
x = np.array([21,27,23,28])
np.mean(x,axis=0)
```

```
x = np.array([21,27,23,28])
x_scaled, mu, st = scaled_array(x)
print('scaled array =',x_scaled,'mean =',mu,'standard devaition =', st)
```
