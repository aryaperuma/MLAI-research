array of all ones

```
def add_one_end_rows(X):
  return np.column_stack((X, np.ones(X.shape[0])))

def best_lin(X,y):
  C = add_one_end_rows(X)
  A = np.matmul(C.T,C)
  print(A)
  r = np.dot(C.T,y)
  print(r)
  z = np.linalg.solve(A,r)
  w = z[:-1]
  b = z[-1]
  return w, b
```
