```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
def f(x,y):
  return x**2+y**2
```

```
xl, xr, dx = -10, 10.01, 0.1
yl, yr, dy = -10, 10.01, 0.1
x = np.arange(xl,xr,dx)
y = np.arange(yl,yr,dy)
f_grid = f(x.reshape(1,len(x)),y.reshape(len(y),1))
```

```
plt.figure(figsize=(4,4)) 
plt.contour(x,y,f_grid,levels=[1,4,9])
plt.xlim(-5,5)
plt.ylim(-5,5)
```
