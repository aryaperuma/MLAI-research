Tensors

```
import numpy as np
import matplotlib.pyplot as plt
```

1 dimensional tensor

```
np.array([1,2])
```

2 dimensional tensor

```
np.array([[1,2],[3,4]])
```

3 dimensional tensor

```
A = np.array([[[1,2,0],[3,4,0]],[[11,12,10],[13,14,10]],[[5,6,0],[7,8,0]]])
print(A)
```

```
A.shape
```

```
A[1,0,1]
```

Images (black and white) are represented with 2 dimensional tensors

```
A = np.zeros((3,4))
A[0,1] = 100
A[1,2] = 200
print(A)
```

```
plt.imshow(A,cmap='Greys',vmin=0,vmax=255)
```

Color images (3 dimensional arrays)

```
A = 255*np.ones((3,4,3))
A[1,2,1] = 100
A[1,2,0] = 100
A[1,2,2] = 0
print(A)
```

```
plt.imshow(A)
```
