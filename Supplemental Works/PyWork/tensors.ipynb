```python
import numpy as np
```

```python
def affine(b,W,X):
  return b + np.matmul(X,W)
```

```python
b = np.array([2,1])
W = np.array([[0,1],[-2,3],[-1,1]])
X = np.array([[0,-1,1],[-1,2,0]])
affine(b,W,X)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def relu(x):
  return np.maximum(0,x)  

def softmax(X):
  return np.exp(X)/(np.sum(np.exp(X),axis=1).reshape(-1,1))   
```

```python
def fully_connected(b,W,activ,X):
  Z = affine(b,W,X)
  if activ == 'sigmoid':
    return sigmoid(Z)
  if activ == 'relu':
    return relu(Z)
  if activ == 'softmax':
    return softmax(Z)
  if activ == 'tanh':
    return np.tanh(Z)
  return Z  
```
