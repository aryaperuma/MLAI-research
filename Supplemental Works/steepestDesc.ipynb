```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x,y):
  return x**2+(y**2-1)**2
  return 3*(x-2*y)**2+(0.2*x+y)**2
  return x**2+y**2 #3*(x-2*y)**2+(0.2*x+y)**2

def f_x(x,y):
  return 2*x
  return 3*2*(x-2*y)+2*0.2*(0.2*x+y)
  return 2*x #3*2*(x-2*y)+2*0.2*(0.2*x+y)

def f_y(x,y):
  return 4*y*(y**2-1)
  return 3*2*(-2)*(x-2*y)+2*(0.2*x+y)    
  return 2*y #3*2*(-2)*(x-2*y)+2*(0.2*x+y)    

def Df(z):
  return np.array([f_x(z[0],z[1]),f_y(z[0],z[1])])  

def steepset(z,r,epochs):
  z_list = np.empty((0,2),float)
  f_list = np.empty(0,float)
  for i in range(epochs):
    z_list = np.append(z_list,[z],axis=0)    
    f_list = np.append(f_list,[f(z[0],z[1])])  
    z = z - r*Df(z)
  z_list = np.append(z_list,[z],axis=0)
  f_list = np.append(f_list,[f(z[0],z[1])])
  return z_list,f_list  
```

```
xl, xr, dx = -10, 10.01, 0.1
yl, yr, dy = -10, 10.01, 0.1
x = np.arange(xl,xr,dx)
y = np.arange(yl,yr,dy)
f_grid = f(x.reshape(1,len(x)),y.reshape(len(y),1))
```

```
epochs = 40
r = 0.1
z = np.array([1,1.5])
z_list, f_list = steepset(z,r,epochs)
```

```
plt.figure(figsize=(4,4)) 
plt.contour(x,y,f_grid,levels=[0.01,0.5,1,2,4])
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.plot(z_list[:,0],z_list[:,1],'ko',markersize=4)
plt.plot(z_list[:,0],z_list[:,1],'k',markersize=4)
plt.plot(z_list[0,0],z_list[0,1],'ro',markersize=4)
```
