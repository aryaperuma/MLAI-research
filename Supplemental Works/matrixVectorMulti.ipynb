```
import numpy as np
```

```
A = np.array([[1,-1,2],[0,3,-2]])
x = np.array([0,1,3])
```

```
np.dot(A,x)
```

```
np.matmul(A,x)
```

```
A = np.array([[0,-1,3],[1,0,1]])
x = np.array([2,1])
print(np.dot(x,A))
print(np.matmul(x,A))
```

```
A = np.array([[3,0,-2],[1,2,0],[3,-1,1]])
b = np.array([1,0,2])
z = np.linalg.solve(A,b)
print(z)
```

```
np.matmul(A,z)
```
