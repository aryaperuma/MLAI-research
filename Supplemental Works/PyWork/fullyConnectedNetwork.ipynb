```python
import numpy as np
```

```python
def pad_ours(p,A):
  C = np.zeros((A.shape[0],A.shape[1]+2*p,A.shape[2]+2*p,A.shape[3]))
  C[:,p:p+A.shape[1],p:p+A.shape[2],:] = A
  return C

def pad(p,A):
  return np.pad(A,((0,0),(p,p),(p,p),(0,0)))
```

```python
A = np.array([[[[1,2,3],[4,5,6]],[[10,20,30],[40,50,60]]],[[[-1,-2,-3],[-4,-5,-6]],[[100,200,300],[400,500,600]]]])
p = 1
B = pad(p,A)
np.pad(A,((0,0),(p,p),(p,p),(0,0))) == B
```
