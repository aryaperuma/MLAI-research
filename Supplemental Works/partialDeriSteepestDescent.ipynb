```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x,y):
  return 3*(x-2*y)**2+(0.2*x+y)**2

def f_x(x,y):
  return 3*2*(x-2*y)+2*0.2*(0.2*x+y)

def f_y(x,y):
  return 3*2*(-2)*(x-2*y)+2*(0.2*x+y)    
```

```
def Df(z):
  return np.array([f_x(z[0],z[1]),f_y(z[0],z[1])])
```

```
epochs = 100
r = 0.05
z = np.array([3,2])
for i in range(epochs):
  z = z - r*Df(z)
  print(z,f(z[0],z[1]))
```
