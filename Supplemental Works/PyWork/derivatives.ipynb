Real valued functions have real numbers as input and give real numbers as output.

For example, $f(x) = x -x^3$

Graph of functions: This is the curve form by the points in the plane of the form $(x,f(x))$

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x):
  return x-x**3
```

```
xl = -1.1
xr = 1.1
dx = 0.01
x = np.arange(xl,xr,dx)
plt.plot(x,f(x),'k')
```

Tangent line: Let $f(x)$ be a real valued function and $x_1$ a number. Think of $x_1$ as fixed. Pick $x_2$ as a second number. Consider the line secant to $f(x)$ through $x_1$ and $x_2$. The tangent line of $f(x)$ at $x_1$ is the line the secant line through $x_1$ and $x_2$ approaches when $x_2$ approaches $x_1$. 

```
xl = -1.1
xr = 1.1
dx = 0.01
x = np.arange(xl,xr,dx)
plt.plot(x,f(x),'k')
x1 = 0.2
x2 = 0.8
plt.plot([x1,x2],[f(x1),f(x2)],'ro')
w = (f(x2)-f(x1))/(x2-x1)
b = f(x1)-w*x1
plt.plot([xl,xr],[w*xl+b,w*xr+b],'r')
x2 = 0.6
plt.plot([x1,x2],[f(x1),f(x2)],'bo')
w = (f(x2)-f(x1))/(x2-x1)
b = f(x1)-w*x1
plt.plot([xl,xr],[w*xl+b,w*xr+b],'b')
x2 = 0.4
plt.plot([x1,x2],[f(x1),f(x2)],'go')
w = (f(x2)-f(x1))/(x2-x1)
b = f(x1)-w*x1
plt.plot([xl,xr],[w*xl+b,w*xr+b],'g')
x2 = 0.2001
plt.plot([x1,x2],[f(x1),f(x2)],'co')
w = (f(x2)-f(x1))/(x2-x1)
b = f(x1)-w*x1
plt.plot([xl,xr],[w*xl+b,w*xr+b],'c')
```

Exponential: $f(x) = e^x$.

```
xl = -2
xr = 2
dx = 0.01
x = np.arange(xl,xr,dx)
plt.plot(x,np.exp(x),'r')
plt.plot([0,0],[0, 7.2],'k')
plt.plot([-2, 2],[0, 0],'k')
plt.text(0.8,5,'$y=e^x$',color = 'r',fontsize=16)
```
