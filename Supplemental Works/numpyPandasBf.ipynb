```
# import the numpy library
import numpy as np
import pandas as pd
```

```
def coefficients(x,y):
  a_11 = (x*x).sum()
  a_12 = x.sum()
  b_1 = (x*y).sum() 
  b_2 = y.sum()
  a_21 = a_12
  a_22 = len(x)  
  return a_11,a_12,b_1,a_21,a_22,b_2

def solution_to_two_linear_equations(a_11,a_12,b_1,a_21,a_22,b_2):
  d = a_11*a_22-a_21*a_12
  x_1 = (a_22*b_1-a_12*b_2)/d
  x_2 = (-a_21*b_1+a_11*b_2)/d 
  return x_1, x_2

def best_line(x,y):
  a_11,a_12,b_1,a_21,a_22,b_2 = coefficients(x,y)
  w, b = solution_to_two_linear_equations(a_11,a_12,b_1,a_21,a_22,b_2)
  return w, b
```
